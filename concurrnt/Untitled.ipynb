```python
from concurrent.futures import ThreadPoolExecutor
```

```python
import time
import secrets
```

```python
import queue
```

```python
def worker_q(q,no):
    print(f'start {no}...')
    while True:
        print(f'q empty {q.empty()}')
        if q.empty():
            print(f'no item ')
#             q.put(None)
            break
        item = q.get()
        print(f'retrive {item}')
        time.sleep(3)
        return no
    
```

```python
def worker(no):
    print(f'start {no}...')
    time.sleep(7)
    print(f'end {no}...')
    return no*2
```

```python
def callback(fu):
    print(f'invoke callback {fu.result()}')
```

```python
p = ThreadPoolExecutor(max_workers=2)
```

```python
def main():
    source = secrets.token_bytes(32)
    print(source)
    q = queue.Queue(maxsize=100)
    for b in source:
        q.put(b)
    with ThreadPoolExecutor(max_workers=2) as e:
        for i in range(1,100):
            fu = e.submit(worker_q,q,i)
            fu.add_done_callback(callback)
            
    
    
```
